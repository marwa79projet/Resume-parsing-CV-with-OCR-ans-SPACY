```python
def ocr_image(image):
    image_gray = cv2.cvtColor(np.array(image), cv2.COLOR_BGR2GRAY)
    text = pytesseract.image_to_string(image_gray)
    return text
```

```python
def clean_text(text):
    review = re.sub(r'(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?"', " ", text)
    review = review.lower()
    review = review.split()
    lm = WordNetLemmatizer()
    review = [lm.lemmatize(word) for word in review if word not in set(stopwords.words("english"))]
    review = " ".join(review)
    return review
```

```python
def get_skills(text):
    doc = nlp_en(text)
    skills = [ent.text for ent in doc.ents if ent.label_ == "SKILL"]
    return skills
```

```python
def unique_skills(skills):
    return list(set(skills))
```

```python
def apply_processing(data):
    data["Clean_Resume"] = data["Resume_str"].apply(clean_text)
    data["skills"] = data["Clean_Resume"].apply(get_skills)
    data["skills"] = data["skills"].apply(unique_skills)
    return data
```

```python
# Sample data
data = pd.DataFrame({
    'Resume_str': ["This is a sample resume.", "Another example of a resume.", "Text data for LDA."]
})

# Define preprocessing function
def preprocess_text(text):
    tokens = text.lower().split()
    return ' '.join(tokens)

# Add the 'Clean_Resume' column
data['Clean_Resume'] = data['Resume_str'].fillna('').apply(preprocess_text)

# Prepare LDA model
docs = data['Clean_Resume'].values
dictionary = corpora.Dictionary(doc.split() for doc in docs)
bow = [dictionary.doc2bow(doc.split()) for doc in docs]

# Train LDA model
num_topics = 4
lda_model = gensim.models.LdaModel(bow, num_topics=num_topics, id2word=dictionary, passes=50, minimum_probability=0)

# Print topics
for topic in lda_model.print_topics(num_topics=num_topics):
    print(topic)

# Visualize topics
pyLDAvis.enable_notebook()
lda_vis = pyLDAvis.gensim_models.prepare(lda_model, bow, dictionary)
pyLDAvis.display(lda_vis)
#pour voir la frequence d'appartion des mots clés dans les differentes categories de CV
```

```python
def delete_cv_by_id(cv_id: int) -> None:
    """
    Supprime un enregistrement de CV de la base de données en fonction de son ID.

    :param cv_id: ID du CV à supprimer.
    """
    try:
        # Rechercher le CV avec l'ID spécifié
        cv_to_delete = session.query(CVData).filter(CVData.id == cv_id).one_or_none()
        if cv_to_delete:
            session.delete(cv_to_delete)
            session.commit()
            print(f"CV avec ID {cv_id} supprimé avec succès.")
        else:
            print(f"Aucun CV trouvé avec l'ID {cv_id}.")
    except Exception as e:
        print(f"Une erreur est survenue: {e}")
        session.rollback()

# Exemple d'utilisation pour supprimer un CV avec ID 1
delete_cv_by_id(1)
```

```python
import re
from sqlalchemy import create_engine, Column, String, Integer, Text
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker

# Configuration de la base de données
DATABASE_URL = 'sqlite:///cv_database.db'
Base = declarative_base()

# Définition du schéma de la base de données
class CVData(Base):
    __tablename__ = 'cv_data'
    id = Column(Integer, primary_key=True, autoincrement=True)
    name = Column(String(255))
    email = Column(String(255))
    phone = Column(String(255))
    skills = Column(Text)
    education = Column(Text)
    experience = Column(Text)

# Création de la base de données
engine = create_engine(DATABASE_URL)
Base.metadata.create_all(engine)
Session = sessionmaker(bind=engine)
session = Session()

# Fonction pour ajouter des données de CV à la base de données
def add_cv_to_db(name, email, phone, skills, education, experience):
    cv_data = CVData(
        name=name,
        email=email,
        phone=phone,
        skills=skills,
        education=education,
        experience=experience
    )
    session.add(cv_data)
    session.commit()

# Exemple d'utilisation de la fonction
add_cv_to_db(
    name='MARWA JEBALI',
    email='jebalimarwa79@gmail.com',
    phone='216-528-044',
    skills='Python, Data Analysis, Machine Learning, Power BI, Tableau, Deep Learning',
    education='Go My Code',
    experience='Expert Data Scientist @ boite DEV'
)

# Exemple de requête pour récupérer des données
def query_cv_data():
    results = session.query(CVData).all()
    for result in results:
        print(f"ID: {result.id}")
        print(f"Name: {result.name}")
        print(f"Email: {result.email}")
        print(f"Phone: {result.phone}")
        print(f"Skills: {result.skills}")
        print(f"Education: {result.education}")
        print(f"Experience: {result.experience}")
        print("-" * 40)  # Séparateur pour chaque CV

query_cv_data()

# Fonctions d'extraction
def extract_name(text):
    name_pattern = r"Name:\s*(.*)"
    match = re.search(name_pattern, text)
    return match.group(1).strip() if match else "Not found"

def extract_email(text):
    email_pattern = r"[\w\.-]+@[\w\.-]+"
    match = re.search(email_pattern, text)
    return match.group(0).strip() if match else "Not found"

def extract_phone(text):
    phone_pattern = r"\(?\b[0-9]{3}[-. )]{0,2}[0-9]{3}[-. ]?[0-9]{4}\b"
    match = re.search(phone_pattern, text)
    return match.group(0).strip() if match else "Not found"

def extract_skills(text):
    skills_pattern = r"Skills:\s*(.*)"
    match = re.search(skills_pattern, text)
    return match.group(1).strip() if match else "Not found"

def extract_education(text):
    education_pattern = r"Education:\s*(.*)"
    match = re.search(education_pattern, text)
    return match.group(1).strip() if match else "Not found"

def extract_experience(text):
    experience_pattern = r"Experience:\s*(.*)"
    match = re.search(experience_pattern, text)
    return match.group(1).strip() if match else "Not found"

# Exemple d'utilisation des fonctions d'extraction
sample_text = """
Name: MARWA JEBALI
Email: jebalimarwa79@gmail.com
Phone: 216-528-044
Skills: Python, Data Analysis, Machine Learning, Power BI, Tableau, Deep Learning
Education: Go My Code
Experience: Expert Data Scientist @ boite DEV
"""

print("Extracted Information:")
print(f"Name: {extract_name(sample_text)}")
print(f"Email: {extract_email(sample_text)}")
print(f"Phone: {extract_phone(sample_text)}")
print(f"Skills: {extract_skills(sample_text)}")
print(f"Education: {extract_education(sample_text)}")
print(f"Experience: {extract_experience(sample_text)}")
```

```python
import os
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from scipy.sparse import hstack
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Chemin vers le répertoire contenant les fichiers de CV
cv_folder_path = '/content/drive/MyDrive/CV base de données'

# Liste pour stocker les textes des CV
cv_texts = []
cv_filenames = []

# Lire les fichiers de CV
for filename in os.listdir(cv_folder_path):
    file_path = os.path.join(cv_folder_path, filename)
    try:
        with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
            cv_texts.append(file.read())
            cv_filenames.append(filename)
    except Exception as e:
        print(f"Erreur lors de la lecture du fichier {filename}: {e}")

# Créer un DataFrame
df = pd.DataFrame({
    'filename': cv_filenames,
    'text': cv_texts
})

print("Aperçu des données :")
print(df.head())

# Assurer que les colonnes ne sont pas vides
df['text'].fillna('', inplace=True)

# Exemple d'extraction de caractéristiques
def extract_features_from_text(text):
    experience = "Experience" if "experience" in text.lower() else ""
    education = "Education" if "education" in text.lower() else ""
    return experience, education

# Extraire les caractéristiques
df['experience'], df['education'] = zip(*df['text'].apply(extract_features_from_text))

print("Aperçu des caractéristiques extraites :")
print(df[['experience', 'education']].head())

# Vérifiez que les colonnes 'experience' et 'education' contiennent des données
df['experience'].replace('', 'No Experience', inplace=True)
df['education'].replace('', 'No Education', inplace=True)

# Vectorisation des textes
vectorizer_exp = CountVectorizer()
X_experience = vectorizer_exp.fit_transform(df['experience'])

vectorizer_edu = CountVectorizer()
X_education = vectorizer_edu.fit_transform(df['education'])

# Combiner les caractéristiques
X = hstack([X_experience, X_education])

# Encoder les étiquettes (exemple d'étiquettes)
categories = ['Developer', 'Manager', 'Scientist']
num_categories = len(categories)
num_samples = len(df)
df['category'] = [categories[i % num_categories] for i in range(num_samples)]

# Encoder les étiquettes
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df['category'])

print(f'X shape: {X.shape}')
print(f'y shape: {y.shape}')

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entraîner le modèle
model = LogisticRegression()
model.fit(X_train, y_train)

# Prédictions
y_pred = model.predict(X_test)

# Évaluation du modèle
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Déterminer les étiquettes présentes dans y_test et y_pred
unique_labels = sorted(set(y_test) | set(y_pred))
target_names = label_encoder.inverse_transform(unique_labels)

# Classification report
report = classification_report(y_test, y_pred, target_names=target_names)
print('Classification Report:')
print(report)
```
